```python
from sympy import init_session
from sympy import *
init_session()
```

```python
R, r, t, l1, l2 = symbols('R r t l_1 l_2', rational=True, positive=True)
# vals = {R: 6, r: 4, t: 2, l1:50, l2:100}
vals = {R: 6, r: 1, t: 2, l1:250, l2:300}
```

```python
a_1 = atan(t/2/R)
a_2 = asin(2*r / sqrt(4*R*R + t**2))
alpha = a_1 + a_2
theta = pi/2 - alpha
```

```python
lb = (R- r * csc(alpha))*csc(alpha)
```

```python
N(lb.subs(vals))
```

```python
N(theta.subs(vals) * 180 / pi)
```

```python
H = (l2 - l1) * cos(theta)
N(H.subs(vals))
```

```python
rl1 = (lb + l1) * sin(theta)
rl2 = (lb + l1 + l1) * sin(theta)
```

```python
N(rl1.subs(vals)), N(rl2.subs(vals))
```

We will take a Cylinder as the working area, r = rl1 and h = H

```python
V = pi * rl1 **2 * H
```

```python
int(N(V.subs(vals)) / 2**3)
```

**Hence, we will have around $10^4$ unit cells.**

```python
V_cone = pi * (rl1 ** 2 + rl2**2) * H / 2
int(N(V_cone.subs(vals)) / 2**3)
```

```python
P, O, c_c = Matrix(symbols('P_0:3', positive=True)), Matrix(symbols('O_0:3', positive=True)), Matrix(symbols('\hat{c}_0:3', positive=True)), 
```

```python
lam = Symbol('\lambda', real=True)
l = Matrix(symbols('l_0:3', real=True))
r = Symbol('r', positive=True)
R = Symbol('R', positive=True)
```

```python
eqs = []
```

```python
eqs.append(c_c.norm()**2 -1)
```

```python
eqs.append(l.norm()**2 -1)
```

```python
Q = P + lam * l
```

```python
eqs.append((Q-O).cross(c_c))
```

```python
eqs.append(R > r)
```

```python
S = O + R * (Q - O) / (Q-O).norm()
```

```python
eqs.append(simplify((S-Q).cross(l).norm()) > r)
```

```python
solve(eqs, k)
```
